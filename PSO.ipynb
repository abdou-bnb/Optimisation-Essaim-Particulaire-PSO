```python
import math
import matplotlib.pyplot as plt
import random

def multip(x,y):
  if x>0 and x<1 :
    return y[:-int(math.floor(len(y)*x))]
  return y*int(abs(x))

random.seed(4)
```

```python
#check_op(a) est une fonction de vérification
#Si la liste a est vide ou si le premier élément de la liste a est de type tuple la fonction renvoie la vitesse
def check_op(a):
  if not a or type(a[0]) is tuple:
    return "vitesse"
  elif all([isinstance(x, int) for x in a]):
    return "vecteur"
  else:
    return None


check_op([0,1,3])
```

```python
#Si les deux operandes sont des vitesses, cette operation est la concatenation de ses deux operandes pour donner une vitesse.
#Si le premier operande est une position et le deuxieme est une vitesse, cette operation est l’equivalent de permutations dans le vecteur de position.
def concat(a, b):
  if check_op(a) == "vitesse" and check_op(b) == "vitesse":
    return a+b
  elif check_op(a) == "vecteur" and check_op(b) == "vitesse":
    if b is None:
      return a
    for ele in b:
      i, j = a.index(ele[0]), a.index(ele[1])
      a[i], a[j] = a[j], a[i]
    return a
  elif check_op(b) == "vecteur" and check_op(a) == "vitesse":
    if a is None:
      return b
    for ele in a:
      i, j = b.index(ele[0]), b.index(ele[1])
      b[i], b[j] = b[j], b[i]
    return b
  return None

concat([4,5,6,7,8],[(6,7),(6,7)])
```

```python
#la fonction substract est utilisé pour mettre a jour les positions des particules dans l'algorithme PSO
def substract(a, b):
  l = a.copy()

  v = []
  for i in range(len(b)-1):
    for j in range(i+1, len(a)):
      if b[i] == l[j]:
        v.append((l[i], b[i]))
        l[i], l[j] = l[j], l[i]
        break
  return v
```

```python
def create_vector(vector_size):
  l = [1]
  for _ in range(vector_size-1):
    x = random.randint(2, vector_size)
    while x in l:
      x = random.randint(2, vector_size)
    l.append(x)
  return l
```

```python
def random_velocity(N): # N refers to number of items in velocity
  velocities = []
  for _ in range(N):
    x1 = random.randint(2, vector_size)
    x2 = random.randint(2, vector_size)
    while (x1, x2) in velocities or x2 == x1:
      x1 = random.randint(2, vector_size)
      x2 = random.randint(2, vector_size)
    velocities.append((x1, x2))
  return velocities

print(create_vector(15))
```

```python



def calculer_cout(x):
  prev_ele = x[0]
  cout = 0
  for ele in x[1:]:
    cout += matrice[prev_ele-1][ele-1]
    prev_ele = ele
  cout += matrice[prev_ele-1][0]
  return cout

```

```python


def next_v(current_v, P, P_best, X): #cest "Vk+1"
  # b1 = b2 = 0.7
  b1 = round(random.uniform(0, 1), 2)
  b2 = round(random.uniform(0, 1), 2)
  ρ = 0.6
  f1 = multip(ρ, current_v)
  # print("f1:", f1)
  f2 = multip(b1, substract(P, X))
  # print("f2:", f2)
  f3 = multip(b2, substract(P_best, X))
  result = concat(concat(f1, f2), f3)
  return result


def next_x(current_x, next_v): #cest "Xk+1"
  return concat(current_x, next_v)


def get_p_best(p):
  min_p = p[0]
  for ele in p[1:]:
    if calculer_cout(min_p) > calculer_cout(ele):
      min_p = ele
  return min_p
```

```python
matrice1 = [
    [0, 6, 6, 1, 4],
    [6, 0, 8, 3, 5],
    [6, 8, 0, 6, 4],
    [1, 3, 6, 0, 2],
    [4, 5, 4, 2, 0]
]



matrice2 = [
    [0,1,1,3,4,5,6,1,7],
    [1,0,5,4,3,6,1,9,2],
    [1,5,0,1,6,1,9,3,7],
    [3,4,1,0,5,9,7,2,1],
    [4,3,6,5,0,9,1,2,1],
    [5,6,1,9,9,0,2,1,7],
    [6,1,9,7,1,2,0,4,5],
    [1,9,3,2,2,1,4,0,7],
    [7,2,7,1,1,7,5,7,0],
]


matrice = matrice2
vector_size = len(matrice)
```

```python
k = 10
vecteurs = []
vitesses = []
print("Vectors:")
for _ in range(k):
  v = create_vector(vector_size)
  
    
  print(v, calculer_cout(v))
  vecteurs.append(v)
  vitesses.append(random_velocity(1))


particules = []
particules_hist = []
print("Particules:")
for _ in range(k):
  p = create_vector(vector_size)
  
    
  print(p, calculer_cout(p))
  particules.append(p)
  particules_hist.append(p)

p_best_0 = create_vector(vector_size)
while p_best_0 in vecteurs + particules:
  p_best_0 = create_vector(5)
print("p_best_0", p_best_0, calculer_cout(p_best_0))

iterations = 1000
counter = 0
pbh = []
new_vitesses = [[] for _ in range(k)]
for j in range(iterations):
  

  new_particules = []
  p_best = get_p_best(particules)
  
    
  new_veteurs = []

  for i, p in enumerate(particules):
    
    v = next_v(vitesses[i], p, p_best, vecteurs[i])
    new_vitesses[i].append(v)
    
    
    x = next_x(vecteurs[i], v)
    
    new_veteurs.append(x)
    
    vitesses[i] = new_vitesses[i].copy()
    new_vitesses[i] = []
    if calculer_cout(p) > calculer_cout(x):
      new_particules.append(x)
      
    else:
      new_particules.append(p)
      

  vecteurs = new_veteurs.copy()
  particules = new_particules.copy()
  
  p_best_0 = p_best
  pbh.append(calculer_cout(p_best_0))
  p_best = get_p_best(particules)
  print("particules=", particules)
  print("calculer_cout(p_best)=", calculer_cout(p_best))
  print("calculer_cout(p_best_0)=", calculer_cout(p_best_0))
  print(p_best_0, calculer_cout(p_best_0))

print("---------------------------------------------------------------------")
print("Number of total iterations to achieve best solution:", j+1)
print(f"P_best={p_best}, cout={calculer_cout(p_best)}")
print(pbh)
print("le cout",min(pbh))
```
